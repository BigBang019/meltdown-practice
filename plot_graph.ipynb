```python
p = "./samples_t"
o_p = "./graph"
for dirname, dir_list, filenames in os.walk(p):
    for dir in dir_list:
        labels = []
        plots = []
        plt.cla()
        for fn in FILENAMES:
            df = pd.read_csv(os.path.join(p, dir, fn + '.txt'), sep=',')
            values = df.iloc[:, 0].values
            plots.append(plt.plot(values, label=fn))
        plt.legend(bbox_to_anchor=(1, 1), bbox_transform=plt.gcf().transFigure
)                
        plt.xlabel('sample')
        plt.ylabel('value')
        plt.title(dir)
        plt.savefig(os.path.join(o_p, dir + '.jpg'), dpi=400, bbox_inches='tight')
```
